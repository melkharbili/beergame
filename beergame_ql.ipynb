# Beer Game Reinforcement Learning


```python
from BPTK_Py.bptk import bptk 
bptk = bptk()
```

```python
bptk.train_scenarios(
    episodes=10,
    scenario_managers=["smBeergameQlOB"],
    scenarios=["train_agents"],
    agents=["controlling"],
    agent_states=["active"],
    agent_properties=["supply_chain_reward"],
    agent_property_types=["total"],
    return_df=False,
    progress_bar=True
)
```
